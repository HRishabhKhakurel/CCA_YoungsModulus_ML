```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data_gbr = pd.read_csv("Feature_importance_gbr_refractory.csv")
data_gbr = data_gbr.iloc[:,1:3]
data_gbr = data_gbr.sort_values(by = ['Importance'],axis = 0, ascending = False)
data_gbr
```

```python
labels = ['V.E.C',r'$\lambda$',r'$\delta$',r'$T_{m}$',r'$\Delta H_{mix}$',r'$a_{m}$',r'$\Delta T_{m}$',r'$\Delta S_{mix}$',r'$\Omega$',r'$\Delta a$',r'$\Delta \chi$']
f, ax = plt.subplots()
sns.barplot(data_gbr.iloc[:,1],data_gbr.iloc[:,0], palette="Blues_d")
ax.set_yticklabels(labels,fontsize = 15)
plt.title('(a)')
plt.xlabel('Feature Importance', fontsize = 12)
plt.ylabel('Features', fontsize = 12)
plt.savefig('plot1.png')
plt.show
```
