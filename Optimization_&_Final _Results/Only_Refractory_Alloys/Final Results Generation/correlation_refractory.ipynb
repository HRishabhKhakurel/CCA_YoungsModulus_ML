```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Youngs_Data_Refractory.csv")
data = data.iloc[:,1:data.shape[1]]
features = data.iloc[:,1:data.shape[1]-1]
```

```python
features
```

```python
correlation = features.corr()
correlation
```

```python
cor = np.array(correlation)
cor
```

```python
labels = [r'$\Delta a$',r'$\Delta T_{m}$',r'$\Delta H_{mix}$',r'$a_{m}$',r'$\lambda$',r'$\delta$',r'$\Omega$',r'$T_{m}$',r'$\Delta \chi$',r'$\Delta S_{mix}$','V.E.C']
```

```python
f, ax = plt.subplots(figsize = (12,12))
sns.heatmap(cor,annot = True)
ax.set_xticklabels(labels,fontsize = 15)
ax.set_yticklabels(labels,fontsize = 15)
plt.savefig('corr_refractory.png')
plt.show
```
